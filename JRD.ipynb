```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#playground 10 Stax X

r = 10
tlde = 0.03
t = 0
nb = 50
ci = 250*nb #clienti iniziali
yrs = 5
mo = 12*yrs
pcon = 3/7 #p consumazione
 
parler = list()
newcj = ci

samap = list()
m = 2.5
y = (8 - m)/2



for nb in range(10,501,10):
  
  #inizializzazione
  newcj = 250*nb
  t = 0
  vv = np.random.normal(y, m , 200)
  a = list([0]*200000)
  z = list([0]*200000)
  b1 = 0
  b2 = 0
  b3 = 0

  for i in range(30):
    a1 = 0 #totale consumato day
    a2 = 0 #totale in day
    a3 = 0 #totale loss day


    for j in range(newcj):

      h = random.randint(0,10)/10

      if h < pcon :

        h2 = random.randint(0,199)

        con = abs(vv[h2]) + 1
        
        if a[j] < con:
          a[j] += r 
          z[j] += r - tlde                
          a2 += r - tlde
          a3 += tlde
      

        a[j] -= con
        z[j] -= con
        a1 += con

        t+= con/100

    #end of day
    b1 += a1
    b2 += a2
    b3 += a3
  
  print(f"con {nb} bar, {newcj} clienti:\nuscite mensili: {round(b1,2)} | entrate mensili: {round(b2,2)} \nlost in transaction: {round(b3,2)} | w/o j: {round(t,2)} \nPortafoglio: {round(sum(a),2)} / {round(sum(z),2)}\n\n")
  parler.append(round(b3,2)/round(t,2))
  samap.append(round(sum(z),2)/round(sum(a),2))

```

```python
plt.plot(parler)
plt.plot(samap)
plt.show()
```
